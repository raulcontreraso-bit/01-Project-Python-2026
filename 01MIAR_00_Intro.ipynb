```python
import subprocess
import sys

def install_if_missing(package):
    try:
        __import__(package)
        print(f"✅ {package} is already installed in your .venv!")
    except ImportError:
        print(f"⚠️ {package} not found. Installing now...")
        # This command runs 'pip install' from inside your notebook
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])
        print(f"✅ {package} installed successfully!")

# List of libraries we want to check
libraries = ['pandas', 'openpyxl', 'matplotlib']

for lib in libraries:
    install_if_missing(lib)
```

```python
import seaborn as sns
df_iris = sns.load_dataset('iris')
print(df_iris.sample(5))  # shows first 5 rows
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import pandas as pd
```

```python
sns.pairplot(data=df_iris, hue='species')
```

```python
#display(df_iris.corr())
display(df_iris.corr(numeric_only=True))
```

```python
#sns.heatmap(df_iris.corr(), square=True, annot=True)
sns.heatmap(df_iris.corr(numeric_only=True), square=True, annot=True)
```

```python
#The result df_pca now contains only the 4 numeric features:
#  sepal_length, sepal_width, petal_length, petal_width
df_pca = df_iris.drop('species', axis=1)
display(df_pca)
```

```python
from sklearn.preprocessing import StandardScaler

x_scaled = StandardScaler()
x_scaled.fit(df_pca)
scaled = x_scaled.transform(df_pca)
print(scaled.shape)
print(scaled)
```

```python
from sklearn.decomposition import PCA
```

```python


n_components = 3
pca = PCA(n_components=n_components)
pca.fit(scaled)
```

```python
x_pca = pca.transform(scaled)
x_pca.shape
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.show()
```

```python
columns = [i for i in pca.explained_variance_ratio_]
columns
```

```python
columns = [f"PC{i+1}" for i, v in enumerate(pca.explained_variance_ratio_)]
columns
```

```python
print(pca.explained_variance_ratio_)
for i, v in enumerate(pca.explained_variance_ratio_):
    print(f"PC{i+1}: {v:.4f} ({v*100:.2f}%)")
```

```python
df = pd.DataFrame(x_pca, columns=columns)
display(df)
df.shape
```

```python
plt.scatter(df.PC1, df.PC2)
plt.show()
```

```python
df_joined = df.join(df_iris['species'], how='inner')
display(df_joined.sample(5))
```

(1) Escribe una función que reciba como entrada una lista con números y devuelva como resultado una lista con los cuadrados de los números contenidos en la lista de entrada.

```python
# Ejercicio 1
def l_cuadrados_2(lista):
    for i in range(len(lista)):       # iterate by index
        lista[i] = pow(lista[i], 2)   # update in place
        #print(lista[i])
    return lista


    
def l_cuadrados(lista):
    t_lista_c = []
    for t_lista in lista:
         #t_lista_c[]  = pow(t_lista,2)
         t_lista_c.append(pow(t_lista, 2))  # build new list
         #print(t_lista,t_lista_c )
    return  t_lista_c

#x = 2
#y = pow(x,2)
#L1 = list() # []
#print(L1)
#print(y)
#print(type(L1))
L1 = [1,2,3,4,6,7,8,9,0]
print(L1)
listas_cuadrados = l_cuadrados_2(L1)

print(listas_cuadrados)
```

# Ejercicio 2: Escribe una función que reciba números como entrada y devuelva la suma de los mismos. La función debe ser capaz de recibir una cantidad indeterminada de números. La función no debe recibir directamente ningún objeto complejo (lista, conjunto, etc.).

```python
# Ejercicio 2

def suma(*numeros):
    r = 0
    if type(numeros) == tuple:
        print(type(numeros)) # Se reciben comoi tuplas
    for Num in numeros:
        r += Num 
    return r

numeros = suma(1, 2, 3,4,5)
print( numeros )

```

# Ejercicio 4: Escribe una función lambda que, al igual que la función desarrollada en el ejercicio anterior, invierta el string recibido como parámetro. Ejemplo: si el string de entrada es 'hola', el resultado será 'aloh'.

```python
x = "Diciembre"
x = lambda x : x + " "
print(x)
```

```python
# Ejercicio 4:
#lambda
#Slicing syntax in Python:
#pythonstring[start : stop : step]
#---------------------------
#Part       Meaning     
#start      where to begin {empty = from the very start/end}
#stop       where to end {empty = go all the way}
#step       how many characters to jump, and in which direction
#The key is step = -1:
#python"hola"[::1]   # → "hola"  — reads forward, every 1 character
#"hola"[::2]   # → "hl"    — reads forward, jumping 2
#"hola"[::-1]  # → "aloh"  — reads BACKWARDS, every 1 character
#A negative step tells Python: "walk through this string from right to left"



    
# Step 1 — define the lambda with [::-1] slicing
invertir = lambda string : string[::-1]

# Step 2 — call it with your word
x = "Diciembre"
print(invertir(x))   # → erbmeiciD ✅
```

Escribe una función que compruebe si un número se encuentra dentro de un rango específico.

```python
def f_rango(numero, inicio, fin):
    o = False
    if numero in range(inicio, fin):
        o = True
    return o

p = f_rango(5, 1, 10)   # is 5 between 1 and 10?
print(p)                 # → True ✅

p = f_rango(15, 1, 10)  # is 15 between 1 and 10?
print(p)                 # → False ✅
```
